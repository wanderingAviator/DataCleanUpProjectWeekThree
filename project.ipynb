# **Imports + Initialization (EXTRACT)**

To begin with, I chose a dataset off of Seaborn: the Dow-Jones Industrial Stock Price Index for United States.

```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
```

```python
stocks = sb.load_dataset('dowjones')
stocks
```

# **Organization + Grouping (TRANSFORM)**

Next, since there wasn't any missing, arbitrary, or otherwise incorrect data, I decided to split up entries by decade...

```python
nineteentens = stocks.loc[(stocks['Date'] >= '1910-01-01')
                     & (stocks['Date'] < '1919-12-31')]

nineteentwenties = stocks.loc[(stocks['Date'] >= '1920-01-01')
                     & (stocks['Date'] < '1929-12-31')]

nineteenthirties = stocks.loc[(stocks['Date'] >= '1930-01-01')
                     & (stocks['Date'] < '1939-12-31')]

nineteenfourties = stocks.loc[(stocks['Date'] >= '1940-01-01')
                     & (stocks['Date'] < '1949-12-31')]

nineteenfifties = stocks.loc[(stocks['Date'] >= '1950-01-01')
                     & (stocks['Date'] < '1959-12-31')]

nineteensixties = stocks.loc[(stocks['Date'] >= '1960-01-01')
                     & (stocks['Date'] < '1969-12-31')]
```

...and by price, in intervals of 100.

```python
nohundreds = stocks.loc[(stocks['Price'] // 100 == 0 )]
onehundreds = stocks.loc[(stocks['Price'] // 100 == 1 )]
twohundreds = stocks.loc[(stocks['Price'] // 100 == 2 )]
threehundreds = stocks.loc[(stocks['Price'] // 100 == 3 )]
fourhundreds = stocks.loc[(stocks['Price'] // 100 == 4 )]
fivehundreds = stocks.loc[(stocks['Price'] // 100 == 5 )]
sixhundreds = stocks.loc[(stocks['Price'] // 100 == 6 )]
sevenhundreds = stocks.loc[(stocks['Price'] // 100 == 7 )]
eighthundreds = stocks.loc[(stocks['Price'] // 100 == 8 )]
ninehundreds = stocks.loc[(stocks['Price'] // 100 == 9 )]
```

# **Data Statistics (ANALYZE)**

All of my analysis was done on price, since the data was logged monthly.

Starting with the decades, I found my Maximum, Minimum, and Averages: Across the whole dataset, and across each sub-group I made....

```python
totalstats = stocks['Price'].describe()
tensstats = nineteentens['Price'].describe()
twentiesstats = nineteentwenties['Price'].describe()
thirtiesstats = nineteenthirties['Price'].describe()
fourtiesstats = nineteenfourties['Price'].describe()
fiftiesstats = nineteenfifties['Price'].describe()
sixtiesstats = nineteensixties['Price'].describe()

totalmean = totalstats['mean']
totalminimum = totalstats['min']
totalmaximum = totalstats['max']
print("Total mean: " + str(totalmean))
print("Total min: " + str(totalminimum))
print("Total max: " + str(totalmaximum))
print()

tensmean = tensstats['mean']
tensminimum =  tensstats['min']
tensmaximum =  tensstats['max']
print("Tens mean: " + str(tensmean))
print("Tens min: " + str(tensminimum))
print("Tens max: " + str(tensmaximum))
print()

twentiesmean = twentiesstats['mean']
twentiesminimum = twentiesstats['min']
twentiesmaximum = twentiesstats['max']
print("Twenties mean: " + str(twentiesmean))
print("Twenties min: " + str(twentiesminimum))
print("Twenties max: " + str(twentiesmaximum))
print()

thirtiesmean = thirtiesstats['mean']
thirtiesminimum = thirtiesstats['min']
thirtiesmaximum = thirtiesstats['max']
print("Thirties mean: " + str(thirtiesmean))
print("Thirties min : " + str(thirtiesminimum))
print("Thirties max: " + str(thirtiesmaximum))
print()

fourtiesmean = fourtiesstats['mean']
fourtiesminimum = fourtiesstats['min']
fourtiesmaximum = fourtiesstats['max']
print("Fourties mean: " + str(fourtiesmean))
print("Fourties min: " + str(fourtiesminimum))
print("Fourties max: " + str(fourtiesmaximum))
print()

fiftiesmean = fiftiesstats['mean']
fiftiesminimum = fiftiesstats['min']
fiftiesmaximum = fiftiesstats['max']
print("Fifties mean: " + str(fiftiesmean))
print("Fifties min: " + str(fiftiesminimum))
print("Fifties max: " + str(fiftiesmaximum))
print()

sixtiesmean = sixtiesstats['mean']
sixtiesminimum = sixtiesstats['min']
sixtiesmaximum = sixtiesstats['max']
print("Sixties mean: " + str(sixtiesmean))
print("Sixties min: " + str(sixtiesminimum))
print("Sixties max: " + str(sixtiesmaximum))
```

...and put them into their own display DataFrame.

```python
statsbydate = pd.DataFrame(
    {
        "Averages" : [totalmean,tensmean,twentiesmean,thirtiesmean,fourtiesmean,fiftiesmean,sixtiesmean],
        "Minimums" : [totalminimum,tensminimum,twentiesminimum,thirtiesminimum,fourtiesminimum,fiftiesminimum,sixtiesminimum],
        "Maximums" : [totalmaximum,tensmaximum,twentiesmaximum,thirtiesmaximum,fourtiesmaximum,fiftiesmaximum,sixtiesmaximum]
    } 
)

statsbydate.index = pd.Series([0,10,20,30,40,50,60])
statsbydate.index.rename("Decade", inplace=True) #"0" is total across the whole dataset, since decade starts in the tens

statsbydate
```

...then putting them into a DataFrame.

```python
statsbyprice = pd.DataFrame(
    {
        "Averages" : [nohundredsmean,onehundredsmean,twohundredsmean,threehundredsmean,fourhundredsmean,fivehundredsmean,sixhundredsmean,sevenhundredsmean,eighthundredsmean,ninehundredsmean],
        "Minimums" : [nohundredsminimum,onehundredsminimum,twohundredsminimum,threehundredsminimum,fourhundredsminimum,fivehundredsminimum,sixhundredsminimum,sevenhundredsminimum,eighthundredsminimum,ninehundredsminimum],
        "Maximums" : [nohundredsmaximum,onehundredsmaximum,twohundredsmaximum,threehundredsmaximum,fourhundredsmaximum,fivehundredsmaximum,sixhundredsmaximum,sevenhundredsmaximum,eighthundredsmaximum,ninehundredsmaximum],
        "Quantity" : [nohundredsquantity,onehundredsquantity,twohundredsquantity,threehundredsquantity,fourhundredsquantity,fivehundredsquantity,sixhundredsquantity,sevenhundredsquantity,eighthundredsquantity,ninehundredsquantity]
    } 
)

statsbyprice.index = pd.Series([0,100,200,300,400,500,600,700,800,900])
statsbyprice.index.rename("Ranges", inplace=True) #between previous value and next

statsbyprice
```

From the DataFrame organized by decade, I can conclude the sixties was the most successful time for Dow-Jones Industrial Stock Prices. On top of having the highest average of all decades, its Maximum and Minimum values are the closest to their respective average of any decade, indicating stability. The 20s (as well as the 30s) saw its average skewed towards its minimum , as to be expected with the Great Depression. Further, across all 6 decades as a whole, this appears to be the case.

As for the DataFrame organized by price, shares saw prices most frequently in the 100-200 range, and least frequently in both the 500-599 range and the 900-999 range, the latter of which is to be expected as stock prices being relatively high is a new phenomena. It's interesting to see a lower-higher patters across the majority of the DataFrame, with the quantities bouncing nearly perfectly back and forth between 40 and 70 per range.

# **Graphing results to confirm them (VISUALIZE)**

Finally, it's time to visualize the data to confirm my observations in the previous steps:

The claims I want to confirm are:
- The 20s and 30s are the only decade ranges to see their average skewed towards their minimum
- There is a peculiar pattern in frequency of prices from $300 - $999

First, the decades:

```python
plt.plot(nineteentens['Date'], nineteentens['Price'], label = '1910s', color='red' )
plt.plot(nineteentwenties['Date'], nineteentwenties['Price'], label = '1920s',color='orange')
plt.plot(nineteenthirties['Date'], nineteenthirties['Price'], label = '1930s',color = 'yellow')
plt.plot(nineteenfourties['Date'], nineteenfourties['Price'], label = '1940s',color = 'green')
plt.plot(nineteenfifties['Date'], nineteenfifties['Price'], label = '1950s',color = 'blue')
plt.plot(nineteensixties['Date'], nineteensixties['Price'], label = '1960s',color = 'purple')
plt.xlabel("Year")
plt.ylabel("Price (Dollars per Share)")
plt.legend()
plt.title("Dow-Jones Indistrial Stock Price Index (By Year)")
plt.show()
```

To look specifically at the ratio of max/mins to average...

```python
plt.bar(np.arange(0,70,10)-3, statsbydate['Averages'].values, width = 3, label = "Averages", color = "red")
plt.bar(np.arange(0,70,10), statsbydate['Maximums'].values, width = 3,label = "Maximum Prices", color = "blue" )
plt.bar(np.arange(0,70,10)+3 , statsbydate['Minimums'].values, width = 3,label = "Minimum Prices", color = "green")

counter = 0
for val in np.arange(statsbydate['Averages'].values.size - 1):
    plt.plot(counter, ((statsbydate['Maximums'].values[val] + statsbydate['Minimums'].values[val]) // 2), marker = "_", color = "cyan")
    counter = counter + 10
plt.plot(counter, ((statsbydate['Maximums'].values[val+1] + statsbydate['Minimums'].values[val+1]) // 2), marker = "_", label = "AVG. of Max + Min", color = "cyan")

counter = -5
for height in statsbydate['Averages'].values:
    plt.text(counter, height, int(height))
    counter = counter + 10

counter = -2
for height in statsbydate['Maximums'].values:
    plt.text(counter, height , int(height))
    counter = counter + 10

counter = 1.5
for height in statsbydate['Minimums'].values:
    plt.text(counter, height , int(height))
    counter = counter + 10

plt.legend()
plt.xlabel("Decade (19XXs) + Total (0)")
plt.ylabel("Price (Dollars per Share)")
plt.title("Dow-Jones Indistrial Stock Price Index (By Year)")

plt.show()
```

It can be a bit tricky to see, but surprisingly, the average price in the 50s also skews towards its minimum, neat!

Now, for the prices: since we already looked at averages, maximums, and minimums, I wanted to look at quantity of occurance of each, instead:

```python
labels = ["\$0 - \$99", "\$100 - \$199", "\$200 - \$299", "\$300 - \$399", "\$400 - \$499", "\$500 - \$599", "\$600 - \$699", "\$700 - \$799", "\$800 - \$899", "\$900 - \$999"]
sizes = statsbyprice['Quantity'].values

graphtitle=plt.title("Dow-Jones Indistrial Stock Price Index (By Price)")
graphtitle.set_ha("center")
plt.gca().axis('equal')
patches, texts = plt.pie(sizes)

graphlabels = ['{0} - {1:1.2f}%'.format(i,j) for i,j in zip(labels, 100.*sizes/sizes.sum())]

plt.legend(patches, graphlabels, bbox_to_anchor=(1,0.5), loc="center right", bbox_transform=plt.gcf().transFigure)
plt.subplots_adjust(left=0.0, bottom=0.1, right=0.7)

plt.show()
```

That's more values from $100-$199 than $300-$999, wow!
